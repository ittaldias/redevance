```python

# 1. Définir les limites de l'espace aérien algérien
latitude_min =  38.95
latitude_max =  39.06 
longitude_min = 4.66919 
longitude_max = 8.00354  

# ... (defining latitude and longitude variables and a_traverse_algerie function)
# Create a boolean Series indicating which SYST_TRAJ_IDs in df_plot_fr are within the specified bounds
is_in_algeria = (df_plot_fr['LATITUDE'] >= latitude_min) & (df_plot_fr['LATITUDE'] <= latitude_max) & (df_plot_fr['LONGITUDE'] >= longitude_min) & (df_plot_fr['LONGITUDE'] <= longitude_max)

# Find the unique SYST_TRAJ_IDs in df_plot_fr that are in Algeria
algeria_syst_traj_ids = df_plot_fr.loc[is_in_algeria, 'SYST_TRAJ_ID'].unique()

# Add 'frontiere_algerienne2' column to df_pln_fr, True if SYST_TRAJ_ID is in algeria_syst_traj_ids, False otherwise
df_pln_fr['frontiere_algerienne'] = df_pln_fr['SYST_TRAJ_ID'].isin(algeria_syst_traj_ids)
```
